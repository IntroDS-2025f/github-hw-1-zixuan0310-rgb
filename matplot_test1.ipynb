````python
import numpy as np
import matplotlib.pyplot as plt

# Âõ∫ÂÆö‰∫ÇÊï∏Á®ÆÂ≠ê‰ª•‰æøÁµêÊûúÂèØÈáçÁèæ
np.random.seed(768)
n = 200

# Êï£‰ΩàÂúñË≥áÊñô
x = np.random.normal(loc=2.0, scale=1.0, size=n)
y = x * 0.5 + np.random.normal(scale=0.8, size=n)
size_var = np.abs(np.random.normal(loc=20, scale=10, size=n))  # Âü∫Á§éÈªûÈù¢Á©ç

# ÁÇ∫‰∫ÜË¶ñË¶∫ÊïàÊûúÔºåÂ∞ç size_var ÂÅöÈÅ©Â∫¶Á∏ÆÊîæ
size_scale = 10  # ‚Üê ÂèØË™øÊï¥ÂÄçÁéáËÆìÂúñÂΩ¢Êõ¥Ê∏ÖÊ•ö
sizes = size_var * size_scale

# Áï´Êï£‰ΩàÂúñÔºåÁî® y ÂÄºÊò†Â∞ÑÈ°èËâ≤
plt.scatter(x, y, s=sizes, c=y, cmap='viridis', alpha=0.7, edgecolor='black')

# Âä†‰∏äÊ®ôÈ°åËàáÈ°èËâ≤Ê¢ù
plt.title('Scatter with Colormap')
cbar = plt.colorbar()
cbar.set_label('y-values')

plt.xlabel('x')
plt.ylabel('y')
plt.show()
````
